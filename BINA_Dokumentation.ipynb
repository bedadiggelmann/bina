### **1.1 Ausgangslage / Problemstellung**
In der politischen Landschaft des Kantons St. Gallen spielt die Wohn-raumthematik eine grosse Rolle, insbesondere im Hinblick auf die sozio-ökonomische Entwicklung und die Lebensqualität der Bevölkerung. Ver-schiedene Medien berichten über Veränderungen auf dem Immobilienmarkt. Beispielsweise veröffentlicht die newhome.ch AG über das Presseportal Informationen über Mietpreissteigerungen und einen sinkenden Leerbe-stand von Wohnungen (https://www.presseportal.ch/de/pm/100096081/100916917). Ersichtlich wird hier, dass der Kanton St. Gallen trotz schweizweiten Steigerungen in den Mietpreisen noch zu den bisher günstigsten Wohnkantonen der Schweiz gehört. Offen ist die Frage, wie sich das aufgrund der Nähe zu Zürich und einer zunehmenden Bevölkerungsdichte noch ändern könnte und welchen Herausforderungen sich der Kanton in Zukunft noch stellen muss.

Damit die Politik für die Zukunft fundierte Entscheidungen treffen kann, ist es wichtig, präzise und verständlich dargestellte Daten zum Wohnraum vorliegen zu haben, auf welche sich politische Entscheidungs-träger verlassen können. In diesem Kontext ist die Verfügbarkeit und Visualisierung von Daten entscheidend. Sie ermöglicht es Politikerinnen und Politikern, Strategien und Entscheidungen zu erarbeiten, die auf realen Gegebenheiten basieren. Dabei kann es sich beispielsweise um die Einzonung von Landwirtschaftszonen handeln oder um das Sprechen von Geldern für zusätzliche Sozialwohnungen.

Das Tagblatt schrieb im März 2024 darüber, dass aufgrund einer Woh-nungsnot in den 1980er Jahren eine Initiative zustande kam, welche für mehr günstigen Wohnraum sorgen sollte
(https://www.tagblatt.ch/ostschweiz/stgallen/lebenskosten-mehr-mieterinnen-und-mieter-sollen-profitieren-stgaller-stadtrat-plant-anpassungen-am-fonds-fuer-guenstige-wohnungen-ld.2590243). Die Frage, welcher sich das Projektteam in dieser Arbeit widmet ist, wie die Daten so aufgearbeitet werden können, damit auf dessen Basis Massnahmen für eine ideale Wohnungssituation auch in der Zukunft getroffen werden kön-nen, damit es nicht wieder zu einer Wohnungsknappheit, wie in den 1980er Jahren kommt.

Aktuell sind die Daten zum Wohnraum im Kanton St. Gallen über verschie-dene Datenquellen verteilt verfügbar und sind oft nicht in ein und der-gleichen Form aufbereitet, was eine schnelle und effektive Analyse zu-lassen würde. Dies erschwert es, einen klaren Überblick über die Wohn-situation zu gewinnen und zeitnahe Massnahmen zu ergreifen.

Die Ausgangslage für die Studierendengruppe besteht somit darin, die vorhandenen Daten zu politisch wichtigen Fragestellungen zu suchen, zu sammeln, auszusortieren, aufzubereiten und zu visualisieren, damit die-se als klare Entscheidungsgrundlage dienen können.
