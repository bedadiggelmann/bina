### **1.1 Ausgangslage / Problemstellung**
In der politischen Landschaft des Kantons St. Gallen spielt die Wohnraumthematik eine grosse Rolle, insbesondere im Hinblick auf die sozioökonomische Entwicklung und die Lebensqualität der Bevölkerung. Verschiedene Medien berichten über Veränderungen auf dem Immobilienmarkt. Beispielsweise veröffentlicht die newhome.ch AG über das Presseportal Informationen über Mietpreissteigerungen und einen sinkenden Leerbestand von Wohnungen (https://www.presseportal.ch/de/pm/100096081/100916917). Ersichtlich wird hier, dass der Kanton St. Gallen trotz schweizweiten Steigerungen in den Mietpreisen noch zu den bisher günstigsten Wohnkantonen der Schweiz gehört. Offen ist die Frage, wie sich das aufgrund der Nähe zu Zürich und einer zunehmenden Bevölkerungsdichte noch ändern könnte und welchen Herausforderungen sich der Kanton in Zukunft noch stellen muss.

Damit die Politik für die Zukunft fundierte Entscheidungen treffen kann, ist es wichtig, präzise und verständlich dargestellte Daten zum Wohnraum vorliegen zu haben, auf welche sich politische Entscheidungsträger verlassen können. In diesem Kontext ist die Verfügbarkeit und Visualisierung von Daten entscheidend. Sie ermöglicht es Politikerinnen und Politikern, Strategien und Entscheidungen zu erarbeiten, die auf realen Gegebenheiten basieren. Dabei kann es sich beispielsweise um die Einzonung von Landwirtschaftszonen handeln oder um das Sprechen von Geldern für zusätzliche Sozialwohnungen.

Das Tagblatt schrieb im März 2024 darüber, dass aufgrund einer Wohnungsnot in den 1980er Jahren eine Initiative zustande kam, welche für mehr günstigen Wohnraum sorgen sollte
(https://www.tagblatt.ch/ostschweiz/stgallen/lebenskosten-mehr-mieterinnen-und-mieter-sollen-profitieren-stgaller-stadtrat-plant-anpassungen-am-fonds-fuer-guenstige-wohnungen-ld.2590243). Die Frage, welcher sich das Projektteam in dieser Arbeit widmet ist, wie die Daten so aufgearbeitet werden können, damit auf dessen Basis Massnahmen für eine ideale Wohnungssituation auch in der Zukunft getroffen werden können, damit es nicht wieder zu einer Wohnungsknappheit, wie in den 1980er Jahren kommt.

Aktuell sind die Daten zum Wohnraum im Kanton St. Gallen über verschiedene Datenquellen verteilt verfügbar und sind oft nicht in ein und dergleichen Form aufbereitet, was eine schnelle und effektive Analyse zulassen würde. Dies erschwert es, einen klaren Überblick über die Wohnsituation zu gewinnen und zeitnahe Massnahmen zu ergreifen.

Die Ausgangslage für die Studierendengruppe besteht somit darin, die vorhandenen Daten zu politisch wichtigen Fragestellungen zu suchen, zu sammeln, auszusortieren, aufzubereiten und zu visualisieren, damit diese als klare Entscheidungsgrundlage dienen können.

## **3 Analyse der gesammelten Daten**

Die Datentabellen wurden vor und auch während der Auswertung analysiert und geprüft, damit sie möglichst vollständig, hochwertig und fehlerfrei sind. Dazu wurden die Tabellen im Excel Format mithilfe der Filterfunktion genauer untersucht oder es wurden bestimmte Werte über einen Python Befehl aufgerufen, um diese zu überprüfen. Folgende Schritte hat das Projektteam bei der Prüfung unternommen:


***Prüfung auf Duplikate:*** Duplikate würden die Ergebnisse verfälschen. Darum erfolgte eine Prüfung auf Duplikate. Es wurden keine Dublikate gefunden.

***Entfernen von irrelevanten Spalten:*** Spalten, die nicht relevant für die Analyse sind oder nur sehr wenige Informationen enthalten, wurden vorzu entfernt, um die Datenmenge zu reduzieren und die Analyse zu erleichtern.

***Kontrolle auf Fehler:*** Falls erkennbare fehlerhafte Einträge in der Datenbank vorhanden sind, sollten diese korrigiert werden. Bei der Kontrolle sind keine Fehler in den Daten aufgefallen.

***Anpassen von Datentypen:*** Die Datentypen wurden laufend so angepasst , dass sie mit den restlichen Daten und den Auswertungsverfahren kompatibel sind.

***Überprüfen der Einheitlichkeit:*** Wenn Daten in unterschiedlichen Einheiten vorliegen, sind diese nur schwer Vergleichbar. Darum sollten sie in die gleiche Einheit umgewandelt werden. Einerseits wurden hierfür aufgrund der Einheit nicht passende Datensätze ausgewechselt und durch Datensätze in der gleichen Einheit (z.B. prozentuale Veränderung statt numerische) ersetzt. Zudem fanden, wenn keine passenden Datensätze in der korrekten Einheit auffindbar waren, zusätzliche Umrechnungen in die passende Einheit statt.

***Bereinigen von fehlenden Werten:*** Die Daten wurden auf fehlende Werte untersucht.
**Fehlende Werte sollten entweder ergänzt oder entfernt werden. Dies hängt von der Anzahl und Bedeutung der fehlenden Werte ab. In einigen Fällen ist es sinnvoll, diese ergänzen zu lassen, in anderen Fällen ist es sinnvoll, die betroffenen Datensätze zu entfernen.
